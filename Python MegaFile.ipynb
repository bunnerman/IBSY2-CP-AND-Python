```python
# MODULE 1

def check_prime(num):
	if num ==1:
		print(" prime no")
		return
	if num==2 or num==3:
		print("prime no")
		return
	if num%2==0 or num%3==0:
		print("Not prime")
		return
	count=5
	for count in range(5,num,count+6):
		if (num % count) == 0 or ((num) % count+2==0):
			print("Not prime")
			return
	print("prime no")

def calculate_factorial(num):
	fact = 1
	if num > 0:
		for count in range(1, num + 1):
			fact *= count
		print("\nFactorial of ", num, " is", fact)
	elif num == 0:
		print("\nFactorial of 0 is 1")
	else:
		print("\nPlease enter a non-negative integer !")
```

```python
class Student:
	# Defining a parameterized constructor having arguments
	def __init__(self, xn, xids, xclg):
		self.n = xn
		self.id = xids
		self.clg = xclg
	def Display_Details(self):
		print("Details")
		print("Name", self.n)
		print("ID", self.id)
		print("College", self.clg)
# Here we create the objects
s1 = Student("Tom", 2023, "MIT")
s2= Student("Bob", 2023, "IIT")
# calling the instance method using the object student
s1.Display_Details()
print()
s2.Display_Details()
```

```python
class BankAccount:
	def __init__(self, xn, xb):
		self.n = xn
		self.b = xb
	def dep(self, amt):
		self.b += amt
	def wth(self, amt):
		self.b = self.b - amt
	def di(self):
		print("Holder:", self.n)
		print("Bal:", self.b)
acc = BankAccount("Tom", 10000)
acc.dep(2000)
acc.wth(1500)
acc.di()
```

```python
class P:
	def f1(self):
		print("Parent")
	# Derived class
class C(P):
	def f2(self):
		print("Child")
# Driver code
obj = C()
obj.f1()
obj.f2()
```

```python
class a:
	def m(self):
		print("In a")
class b(a):
	def m(self):
		print("In b")
		super().m()
class c(a):
	def m(self):
		print("In c")
		super().m()
class d(b, c):
	def m(self):
		print("In d")
		super().m()
obj = d()
obj.m()
```

```python
class School:
	def func1(self):
		print("This function is in school.")
# Derived class 1 (Single Inheritance)
class Student1(School):
	def func2(self):
		print("This function is in student 1.")
# Derived class 2 (Another Single Inheritance)
class Student2(School):
	def func3(self):
		print("This function is in student 2.")
# Derived class 3 (Multiple Inheritance)
class Student3(Student1, School):
	def func4(self):
		print("This function is in student 3.")
# Driver code
obj = Student3()
obj.func1()
obj.func2()
```
